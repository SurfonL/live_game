# BINARY MASK

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np



img_dir = 'monsters/Rabbit/rabbit.jpg'
mask_dir = img_dir.replace('rabbit', 'rabbit_mask')


mask = cv2.imread(img_dir,0)
# th,img = cv2.threshold(img,0,255,cv2.THRESH_OTSU)

th = 50
mask[mask<th] = 0
mask[mask>=th] = 255


cv2.imwrite(mask_dir, mask)

plt.imshow(mask)
plt.show()
```
